# Q10. Plot violin plot between above respective 13 categories of unique year month combinations Month_Year ( 1_2019 , 2_2019 â€¦. 1_2020) along side the number of orders being placed. Write down the inferences according to the graph. (minimum 5) 

```python
df.head()
```

```python
de=df
```

```python
de.head()
```

```python
de['Month_Year']=df['month'].astype(str)+"_"+df['year'].astype(str)
```

```python
de.head()
```

```python
de['Month_Year'].unique()
```

```python
sns.violinplot(data=de,x='quantity',y='Month_Year',width=1.5,gridsize=400,dodge=False)
plt.show()
plt.tight_layout()
```

Inferences :-

1. Jan of 2020 which is 1_2020 is having a narrow belly that means the density of the data is less compared to the other months in 2019

2. The steep of the peak in 7_2019(July of 2019) and 6_2019(June of 2019) is very high which means we have a very strong       positive outlier

3. 1_2020(Jan of 2020) is having a narrow peak compared to others although it is steep so it has a weak outlier compared to the other months of 2019

4. The  IQR of the points except 1_2020 seems quite similar values since we can infer that the size of the plot is similar from the small graph

5. Since we see steep rise in the outliers we can say that the 'quantity' is more focussed towards the lower area that is 1 to 2 quantiy mostly for all the months in both the years. Mostly 1 quantiy is being sold.

Conclusion: We can say that mostly 1 quantity is being sold all over the time with a very small percentage of other values
